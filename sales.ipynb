```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('sales_data_sample.csv', encoding='unicode_escape')
df.head()
```

```python
df_drop = ['ADDRESSLINE1', 'ADDRESSLINE2', 'POSTALCODE', 'CITY', 'TERRITORY', 'PHONE']
df.drop(columns=df_drop, inplace=True, errors='ignore')
df.info()
```

```python
df.drop(columns=['ORDERDATE', 'STATUS', 'MONTH_ID', 'QTR_ID', 'YEAR_ID'], inplace=True, errors='ignore')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

def convert_categories(col):
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].values)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
```

```python
from sklearn.cluster import KMeans
wcss = []
for k in range(1, 15):
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(df[numerical_cols])  # Assuming 'df[numerical_cols]' contains the relevant data
    wcss.append(kmeans.inertia_)
k = list(range(1,15))
plt.plot(k,wcss)
plt.xlabel('Clusters')
plt.ylabel('scores')
plt.title('Finding right number of clusters')
plt.grid()
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
data = load_iris()
X = data.data  
y = data.target  
parameters_grid = {"n_neighbors": np.arange(0, 50)}
knn = KNeighborsClassifier()
knn_GSV = GridSearchCV(estimator=knn, param_grid=parameters_grid, cv=5)
knn_GSV.fit(X, y)
print("Best Param", knn_GSV.best_params_)
print("Best score", knn_GSV.best_score_)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, fbeta_score
def model_evaluation(y_test, y_pred, model_name):
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='weighted')
    rec = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    f2 = fbeta_score(y_test, y_pred, beta=2, average='weighted')  # Adjust beta as needed

    results = pd.DataFrame([[model_name, acc, prec, rec, f1, f2]],
                          columns=["Model", "Accuracy", "Precision", "Recall",
                                   "F1 Score", "F2 Score"])
    results = results.sort_values(["Precision", "Recall", "F2 Score"], ascending=[False, False, False])
    return results
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_auc_score, roc_curve

data = load_iris()
X = data.data  
y = (data.target == 1).astype(int) 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
KNeighborsClassifier()
y_pred_proba = knn.predict_proba(X_test)[:, -1] 
fpr, tpr, threshold = roc_curve(y_test, y_pred_proba)
roc_auc = roc_auc_score(y_test, y_pred_proba)
print(f"AUC Score: {roc_auc}")
plt.plot(fpr, tpr, label=f"KNN (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], 'k--')  
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('KNN ROC Curve')
plt.legend(loc='lower right')
plt.show()
```
